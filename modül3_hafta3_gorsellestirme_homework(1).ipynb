# üéì Titanic Veri Seti G√∂rselle≈ütirme Uygulamalarƒ±
Her sorunun altƒ±nda cevabƒ±nƒ±zƒ± kodlayabilirsiniz.
Grafiklerinizi yorumlamanƒ±z ve grafikten √ßƒ±karƒ±mlarƒ±nƒ±zƒ± yazmanƒ±z gerekmektedir.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = sns.load_dataset("titanic")
df.head()
```

## üü¢ Kolay Seviye

### ‚ùì Soru: Cinsiyete g√∂re yolcu sayƒ±sƒ±nƒ± g√∂rselle≈ütirin. (√∂rn. countplot veya bar chart)

```python
sex_counts = df['sex'].value_counts()
sex_counts.index = ['Erkek', 'Kadƒ±n'] # Rename index to Turkish

plt.figure(figsize=(12,8)) # Grafik boyutu
plt.bar(sex_counts.index, sex_counts.values, color=['blue', 'hotpink'], width=0.3) # Bar grafiƒüi olu≈ütur
plt.title("Cinsiyet Daƒüƒ±lƒ±mƒ±", fontsize=14) # Ba≈ülƒ±k
plt.xlabel("Cinsiyeti") # X ekseni etiketi
plt.ylabel("Yolcu Sayƒ±sƒ±") # Y ekseni etiketi
plt.tight_layout() # Grafik d√ºzenini sƒ±kƒ±≈ütƒ±r
plt.show()
```

### ‚ùì Soru: "Survived" deƒüi≈ükeninin daƒüƒ±lƒ±mƒ±nƒ± √ßiziniz.

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='survived', hue='survived', palette=['darkred', 'lightgreen'])
plt.title("Hayatta Kalma Daƒüƒ±lƒ±mƒ±", fontsize=14)
plt.xlabel("Hayatta Kalma")
plt.ylabel("Yolcu Sayƒ±sƒ±")
plt.xticks([0, 1], ["√ñl√º", "Hayatta"])
plt.show()
```

### ‚ùì Soru: "Pclass" deƒüi≈ükeninin histogramƒ±nƒ± veya √ßubuk grafiƒüini olu≈üturun.

```python
pclass_counts = df['pclass'].value_counts().sort_index()

plt.figure(figsize=(12, 8))
plt.bar(pclass_counts.index, pclass_counts.values, color=['#89CFF0', '#FED8B1', 'lightgreen'])
plt.title("Pclass Daƒüƒ±lƒ±mƒ±", fontsize=14)
plt.xlabel("Pclass")
plt.ylabel("Yolcu Sayƒ±sƒ±")
plt.xticks([1, 2, 3])
plt.show()
```

### ‚ùì Soru: "Age" deƒüi≈ükeninin histogramƒ±nƒ± √ßiziniz.

```python
sns.histplot(df["age"], bins=30)
plt.title("Ya≈ü Daƒüƒ±lƒ±mƒ±") # Ba≈ülƒ±k ekle
plt.xlabel("Ya≈ü") # X eksenine etiket ekle
plt.ylabel("Frekans") # Y eksenine etiket ekle
plt.show() # Grafiƒüi g√∂ster
```

### ‚ùì Soru: "Fare" deƒüi≈ükeninin histogramƒ±nƒ± √ßizip a√ßƒ±klayƒ±nƒ±z.

```python
# Cevabƒ±nƒ±zƒ± buraya
sns.histplot(df["fare"], bins=50)
plt.title("√úcret Daƒüƒ±lƒ±mƒ±") # Ba≈ülƒ±k ekle
plt.xlabel("√úcret") # X eksenine etiket ekle
plt.ylabel("Frekans") # Y eksenine etiket ekle
plt.show() # Grafiƒüi g√∂ster
```

## üü° Orta Seviye

### ‚ùì Soru: Cinsiyete g√∂re ya≈ü daƒüƒ±lƒ±mƒ±nƒ± kar≈üƒ±la≈ütƒ±rƒ±n. (√∂rn. boxplot, violinplot, kdeplot)

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
# Python program to illustrate
# violinplot using inbuilt data-set
# given in seaborn

# importing the required module

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set

sns.violinplot(x ='sex', y ='age', data = df, linewidth = 4, color="#C3B091")
```

### ‚ùì Soru: Pclass'a g√∂re √ºcret daƒüƒ±lƒ±mƒ±nƒ± inceleyin.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='pclass', y='fare', hue='pclass', palette=["red","blue","yellow"])
plt.title('Pclass\'a G√∂re √úcret Daƒüƒ±lƒ±mƒ±')
plt.xlabel('Pclass')
plt.ylabel('√úcret')
plt.show()
```

### ‚ùì Soru: Survived‚Äôa g√∂re ya≈ü daƒüƒ±lƒ±mƒ±nƒ± inceleyin.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
plt.figure(figsize=(12,8))
sns.violinplot(data=df, x='survived', y='age', hue='survived', palette=["brown","yellow"])
plt.ylabel('Ya≈ü')
plt.xlabel('Hayatta Kalma Durumu')
plt.xticks([0, 1], ["√ñl√º", "Hayatta"])
plt.show()
```

### ‚ùì Soru: Cinsiyet ve sƒ±nƒ±fa g√∂re hayatta kalma oranlarƒ±nƒ± barplot ile kar≈üƒ±la≈ütƒ±rƒ±n.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
plt.figure(figsize=(12,8))
sns.countplot(data=df, x='sex', hue='pclass', palette=["red","blue","yellow"])
```

### ‚ùì Soru: FacetGrid ile cinsiyete g√∂re ya≈ü histogramlarƒ±nƒ± ayrƒ± ayrƒ± g√∂sterin.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
dt = sns.FacetGrid(df, col="sex")
dt.map(plt.hist, "age")
```

### ‚ùì Soru: "Fare" ve "Age" deƒüi≈ükenleri arasƒ±nda scatterplot olu≈üturun.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
sns.scatterplot(data=df,y='fare',x='age')
```

### ‚ùì Soru: Korelasyon matrisi ve heatmap ile deƒüi≈ükenler arasƒ±ndaki ili≈ükiyi g√∂rselle≈ütirin.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="magma")
plt.title("Korelasyon Matrisi") # Ba≈ülƒ±k ekle
plt.show() # Grafiƒüi g√∂ster
```

## üî¥ ƒ∞leri Seviye

### ‚ùì Soru: "Age" deƒüi≈ükeninin √ßarpƒ±klƒ±k (skewness) ve basƒ±klƒ±ƒüƒ±nƒ± (kurtosis) hesaplayƒ±n ve yorumlayƒ±n.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
# SciPy version
from scipy.stats import skew
from scipy.stats import kurtosis

skew(df['age'])
```

Ya≈ü deƒüi≈ükeninde bo≈ü veriler olduƒüu i√ßin √ßarpƒ±klƒ±ƒüƒ±nƒ± hesaplayamƒ±yoruz o y√ºzden temizlik vakti
üßº‚ú®üßπüß∫ü´ßüßΩüß¥üíßüßº‚ú®

```python
notNullAge = df['age'].dropna()
print(skew(notNullAge))
print(kurtosis(notNullAge))
```

### ‚ùì Soru: Seaborn pairplot ile sayƒ±sal deƒüi≈ükenleri inceleyin. Hayatta kalma durumunu hue parametresiyle ayƒ±rƒ±n.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
numerical_df = df.select_dtypes(include=np.number)
sns.pairplot(data=numerical_df, hue='survived')
```

### ‚ùì Soru: "Sex", "Pclass" ve "Survived" deƒüi≈ükenlerini i√ßeren gruplu bir violinplot olu≈üturun.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
plt.figure(figsize=(12,8))
sns.violinplot(data=df, x='survived', y='pclass', hue='sex', palette=["orange","purple"])
```

### ‚ùì Soru: "Embarked" deƒüi≈ükenine g√∂re fare daƒüƒ±lƒ±mƒ±nƒ± inceleyin.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
plt.figure(figsize=(12,8))
sns.boxplot(data=df, x='embarked', y='fare')
```

### ‚ùì Soru: Plotly ile cinsiyet ve sƒ±nƒ±fa g√∂re interaktif bir histogram olu≈üturun.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
import plotly.express as px

fig = px.histogram(df, x='sex', color='pclass',

                   labels={'sex':'Cinsiyet', 'pclass':'Yolcu Sƒ±nƒ±fƒ±'})

fig.show()
```

### ‚ùì Soru: Eksik veri analizi yaparak eksik verileri heatmap ile g√∂rselle≈ütirin.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
plt.figure(figsize=(12,8))
sns.heatmap(df.isnull(), cbar=False, cmap='magma')
```

### ‚ùì Soru: Cinsiyet ve ya≈ü arasƒ±ndaki ili≈ükiyi scatterplot ile g√∂sterip renkleri hayatta kalma durumuna g√∂re ayarlayƒ±n.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, y='age', x='sex', hue='survived')
```

### ‚ùì Soru: Matplotlib ile √∂zelle≈ütirilmi≈ü bir bar grafiƒüi olu≈üturun (renk, kalƒ±nlƒ±k, grid, label).

```python
plt.figure(figsize=(12, 8)) # Grafik boyutunu ayarlama

plt.bar(pclass_counts.index, pclass_counts.values,color=['skyblue', 'lightgreen', 'darkred'], width=0.5)


plt.title('Pclass Yolcu Sayƒ±sƒ± Grafiƒüi', fontsize=16, fontweight='bold')
plt.xlabel('Yolcu Sƒ±nƒ±fƒ± (Pclass)', fontsize=12)
plt.ylabel('Yolcu Sayƒ±sƒ±', fontsize=12)
plt.xticks(pclass_counts.index, ['1. Sƒ±nƒ±f', '2. Sƒ±nƒ±f', '3. Sƒ±nƒ±f'], fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```
