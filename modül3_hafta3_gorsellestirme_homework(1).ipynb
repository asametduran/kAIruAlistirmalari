# 🎓 Titanic Veri Seti Görselleştirme Uygulamaları
Her sorunun altında cevabınızı kodlayabilirsiniz.
Grafiklerinizi yorumlamanız ve grafikten çıkarımlarınızı yazmanız gerekmektedir.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = sns.load_dataset("titanic")
df.head()
```

## 🟢 Kolay Seviye

### ❓ Soru: Cinsiyete göre yolcu sayısını görselleştirin. (örn. countplot veya bar chart)

```python
sex_counts = df['sex'].value_counts()
sex_counts.index = ['Erkek', 'Kadın'] # Rename index to Turkish

plt.figure(figsize=(12,8)) # Grafik boyutu
plt.bar(sex_counts.index, sex_counts.values, color=['blue', 'hotpink'], width=0.3) # Bar grafiği oluştur
plt.title("Cinsiyet Dağılımı", fontsize=14) # Başlık
plt.xlabel("Cinsiyeti") # X ekseni etiketi
plt.ylabel("Yolcu Sayısı") # Y ekseni etiketi
plt.tight_layout() # Grafik düzenini sıkıştır
plt.show()
```

### ❓ Soru: "Survived" değişkeninin dağılımını çiziniz.

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='survived', hue='survived', palette=['darkred', 'lightgreen'])
plt.title("Hayatta Kalma Dağılımı", fontsize=14)
plt.xlabel("Hayatta Kalma")
plt.ylabel("Yolcu Sayısı")
plt.xticks([0, 1], ["Ölü", "Hayatta"])
plt.show()
```

### ❓ Soru: "Pclass" değişkeninin histogramını veya çubuk grafiğini oluşturun.

```python
pclass_counts = df['pclass'].value_counts().sort_index()

plt.figure(figsize=(12, 8))
plt.bar(pclass_counts.index, pclass_counts.values, color=['#89CFF0', '#FED8B1', 'lightgreen'])
plt.title("Pclass Dağılımı", fontsize=14)
plt.xlabel("Pclass")
plt.ylabel("Yolcu Sayısı")
plt.xticks([1, 2, 3])
plt.show()
```

### ❓ Soru: "Age" değişkeninin histogramını çiziniz.

```python
sns.histplot(df["age"], bins=30)
plt.title("Yaş Dağılımı") # Başlık ekle
plt.xlabel("Yaş") # X eksenine etiket ekle
plt.ylabel("Frekans") # Y eksenine etiket ekle
plt.show() # Grafiği göster
```

### ❓ Soru: "Fare" değişkeninin histogramını çizip açıklayınız.

```python
# Cevabınızı buraya
sns.histplot(df["fare"], bins=50)
plt.title("Ücret Dağılımı") # Başlık ekle
plt.xlabel("Ücret") # X eksenine etiket ekle
plt.ylabel("Frekans") # Y eksenine etiket ekle
plt.show() # Grafiği göster
```

## 🟡 Orta Seviye

### ❓ Soru: Cinsiyete göre yaş dağılımını karşılaştırın. (örn. boxplot, violinplot, kdeplot)

```python
# Cevabınızı buraya yazınız
# Python program to illustrate
# violinplot using inbuilt data-set
# given in seaborn

# importing the required module

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set

sns.violinplot(x ='sex', y ='age', data = df, linewidth = 4, color="#C3B091")
```

### ❓ Soru: Pclass'a göre ücret dağılımını inceleyin.

```python
# Cevabınızı buraya yazınız
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='pclass', y='fare', hue='pclass', palette=["red","blue","yellow"])
plt.title('Pclass\'a Göre Ücret Dağılımı')
plt.xlabel('Pclass')
plt.ylabel('Ücret')
plt.show()
```

### ❓ Soru: Survived’a göre yaş dağılımını inceleyin.

```python
# Cevabınızı buraya yazınız
plt.figure(figsize=(12,8))
sns.violinplot(data=df, x='survived', y='age', hue='survived', palette=["brown","yellow"])
plt.ylabel('Yaş')
plt.xlabel('Hayatta Kalma Durumu')
plt.xticks([0, 1], ["Ölü", "Hayatta"])
plt.show()
```

### ❓ Soru: Cinsiyet ve sınıfa göre hayatta kalma oranlarını barplot ile karşılaştırın.

```python
# Cevabınızı buraya yazınız
plt.figure(figsize=(12,8))
sns.countplot(data=df, x='sex', hue='pclass', palette=["red","blue","yellow"])
```

### ❓ Soru: FacetGrid ile cinsiyete göre yaş histogramlarını ayrı ayrı gösterin.

```python
# Cevabınızı buraya yazınız
dt = sns.FacetGrid(df, col="sex")
dt.map(plt.hist, "age")
```

### ❓ Soru: "Fare" ve "Age" değişkenleri arasında scatterplot oluşturun.

```python
# Cevabınızı buraya yazınız
sns.scatterplot(data=df,y='fare',x='age')
```

### ❓ Soru: Korelasyon matrisi ve heatmap ile değişkenler arasındaki ilişkiyi görselleştirin.

```python
# Cevabınızı buraya yazınız
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="magma")
plt.title("Korelasyon Matrisi") # Başlık ekle
plt.show() # Grafiği göster
```

## 🔴 İleri Seviye

### ❓ Soru: "Age" değişkeninin çarpıklık (skewness) ve basıklığını (kurtosis) hesaplayın ve yorumlayın.

```python
# Cevabınızı buraya yazınız
# SciPy version
from scipy.stats import skew
from scipy.stats import kurtosis

skew(df['age'])
```

Yaş değişkeninde boş veriler olduğu için çarpıklığını hesaplayamıyoruz o yüzden temizlik vakti
🧼✨🧹🧺🫧🧽🧴💧🧼✨

```python
notNullAge = df['age'].dropna()
print(skew(notNullAge))
print(kurtosis(notNullAge))
```

### ❓ Soru: Seaborn pairplot ile sayısal değişkenleri inceleyin. Hayatta kalma durumunu hue parametresiyle ayırın.

```python
# Cevabınızı buraya yazınız
numerical_df = df.select_dtypes(include=np.number)
sns.pairplot(data=numerical_df, hue='survived')
```

### ❓ Soru: "Sex", "Pclass" ve "Survived" değişkenlerini içeren gruplu bir violinplot oluşturun.

```python
# Cevabınızı buraya yazınız
plt.figure(figsize=(12,8))
sns.violinplot(data=df, x='survived', y='pclass', hue='sex', palette=["orange","purple"])
```

### ❓ Soru: "Embarked" değişkenine göre fare dağılımını inceleyin.

```python
# Cevabınızı buraya yazınız
plt.figure(figsize=(12,8))
sns.boxplot(data=df, x='embarked', y='fare')
```

### ❓ Soru: Plotly ile cinsiyet ve sınıfa göre interaktif bir histogram oluşturun.

```python
# Cevabınızı buraya yazınız
import plotly.express as px

fig = px.histogram(df, x='sex', color='pclass',

                   labels={'sex':'Cinsiyet', 'pclass':'Yolcu Sınıfı'})

fig.show()
```

### ❓ Soru: Eksik veri analizi yaparak eksik verileri heatmap ile görselleştirin.

```python
# Cevabınızı buraya yazınız
plt.figure(figsize=(12,8))
sns.heatmap(df.isnull(), cbar=False, cmap='magma')
```

### ❓ Soru: Cinsiyet ve yaş arasındaki ilişkiyi scatterplot ile gösterip renkleri hayatta kalma durumuna göre ayarlayın.

```python
# Cevabınızı buraya yazınız
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, y='age', x='sex', hue='survived')
```

### ❓ Soru: Matplotlib ile özelleştirilmiş bir bar grafiği oluşturun (renk, kalınlık, grid, label).

```python
plt.figure(figsize=(12, 8)) # Grafik boyutunu ayarlama

plt.bar(pclass_counts.index, pclass_counts.values,color=['skyblue', 'lightgreen', 'darkred'], width=0.5)


plt.title('Pclass Yolcu Sayısı Grafiği', fontsize=16, fontweight='bold')
plt.xlabel('Yolcu Sınıfı (Pclass)', fontsize=12)
plt.ylabel('Yolcu Sayısı', fontsize=12)
plt.xticks(pclass_counts.index, ['1. Sınıf', '2. Sınıf', '3. Sınıf'], fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```
