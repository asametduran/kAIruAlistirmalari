# 📘 Veri Bilimi Nedir?
### 📖 Tanım:

**Veri Bilimi**, yapılandırılmış veya yapılandırılmamış verilerden anlamlı içgörüler çıkarma sürecidir. Bu süreçte istatistik, programlama, alan bilgisi ve problem çözme becerileri birlikte kullanılır.

Kısaca:

> "Veriden değer üretme bilimidir."

---

## 🧪 Veri Biliminin Temel Amaçları:

1. **Descriptive (Betimleyici) Analiz:**
   "Neler oldu?" sorusuna cevap arar.
   Örnek: Satışlar bu çeyrekte %10 arttı.

2. **Diagnostic (Tanısal) Analiz:**
   "Neden oldu?" sorusunu araştırır.
   Örnek: Satışlar neden düştü?

3. **Predictive (Tahminsel) Analiz:**
   "Gelecekte ne olacak?"
   Örnek: Önümüzdeki ay 1000 birim satış bekleniyor.

4. **Prescriptive (Yönlendirici) Analiz:**
   "Ne yapmalıyız?"
   Örnek: Hangi kampanya satışları artırır?

# 👨‍🔬 Veri Bilimci Ne Yapar?

### 👔 Genel Rol:

Veri bilimcisi, veriyle uğraşan, analiz eden, model kuran ve bunları **iş değeri üretecek şekilde** yorumlayan kişidir.

---

## 🔁 Veri Bilimcinin Tipik İş Akışı (Pipeline):

1. **Problem Tanımı:**

   * İş problemi nedir? (ör: Müşteri neden terk ediyor?)
   * Veri bilimi bu probleme nasıl katkı sağlar?

2. **Veri Toplama (Data Acquisition):**

   * Kaynak: veritabanı, API, dosya, anket...
   * SQL, web scraping, sensor data, vs.

3. **Veri Temizleme (Data Cleaning):**

   * Eksik değerleri doldurma
   * Aykırı değerleri tespit etme
   * Veri tipi dönüşümleri

4. **Keşifsel Veri Analizi (EDA):**

   * Veriyi anlama ve ilk içgörüler
   * Görselleştirme
   * Grup bazlı analizler

5. **Özellik Mühendisliği (Feature Engineering):**

   * Yeni değişkenler üretme
   * Kategorik değişkenleri kodlama
   * Değer ölçekleme (scaling)

6. **Modelleme (Makine Öğrenmesi):**

   * Regresyon, sınıflandırma, kümeleme...
   * Model seçimi, eğitim ve değerlendirme

7. **Model Değerlendirme:**

   * Accuracy, Precision, Recall, F1-score, ROC-AUC
   * Cross-validation

8. **Modelin Dağıtımı (Deployment):**

   * API, dashboard, otomatik pipeline
   * Streamlit, FastAPI, Flask

9. **Raporlama ve İşe Etki:**

   * Veri hikayesi anlatımı
   * Karar vericilere sunum
   * İş stratejilerine katkı

## Değişken Türleri
Veri biliminde değişkenleri doğru tanımlamak analizlerin doğruluğu açısından çok önemlidir. Pandas'ta dtypes ile görülen veri tipleri (object, int, float vs.) yeterli değildir. Bunun yerine anlamına göre sınıflandırma gerekir.

🔹 Kategorik Değişken (Categorical Variables):
Sayısal görünse bile, sınıf belirten değişkenlerdir.

Örnek: Sex, Embarked, Pclass, Survived


🔹 Sayısal Değişken (Numerical Variables):
Gerçekten miktar ifade eder.

Üzerinde matematiksel işlemler yapılabilir.

Örnek: Age, Fare

🔹 Kategorik ama Sayısal Görünen (num_but_cat):
Dtype'ı int veya float olabilir ama anlamı kategoriktir.

Genellikle 10'dan az farklı değer içerir.

Örnek: Pclass, Survived

🔹 Kardinal Değişken (High Cardinality Categorical):
Çok fazla kategori içerir (20+ gibi).

Genellikle object tipindedir.

Örnek: Name, Ticket, Cabin

Kodlamalarda sorun çıkarabilir; genellikle metinsel ve tekil oldukları için analiz dışında bırakılır veya özel işleme alınır.

## TİTANİC VERİ SETİNİ TANIYALIM

1. **PassengerId**: Her yolcuya atanan benzersiz bir kimlik numarası. Bu değişken, yolcuları ayırt etmek için kullanılır ve analizde genellikle bir indeks olarak işlev görür.

2. **Survived**: Yolcunun hayatta kalıp kalmadığını gösteren hedef değişken.
   - 0: Hayatta kalmadı
   - 1: Hayatta kaldı

3. **Pclass**: Yolcunun bilet sınıfı, sosyo-ekonomik durumun bir göstergesi olarak kabul edilir.
   - 1: Birinci sınıf
   - 2: İkinci sınıf
   - 3: Üçüncü sınıf

4. **Name**: Yolcunun tam adı. Bu değişken, genellikle unvanlar (örneğin, Mr., Mrs., Miss) gibi ek bilgiler çıkarmak için kullanılır.

5. **Sex**: Yolcunun cinsiyeti.
   - male: Erkek
   - female: Kadın

6. **Age**: Yolcunun yaşı. Bazı durumlarda eksik (NaN) olabilir, bu yüzden veri ön işleme sırasında eksik değerlerin doldurulması gerekebilir.

7. **SibSp**: Yolcunun gemideki kardeş/eş sayısı. Bu, yolcunun gemideki yakın aile üyelerinin (kardeş, üvey kardeş, eş) sayısını ifade eder.

8. **Parch**: Yolcunun gemideki ebeveyn/çocuk sayısı. Bu, yolcunun gemideki ebeveyn veya çocuklarının sayısını belirtir.

9. **Ticket**: Bilet numarası. Her yolcu için benzersiz bir bilet numarasıdır, ancak bazı yolcular aynı bilet numarasını paylaşabilir (örneğin, aynı aileden olanlar).

10. **Fare**: Bilet ücreti. Yolcunun ödediği miktarı gösterir ve sosyo-ekonomik durumla ilişkilendirilebilir.

11. **Cabin**: Yolcunun kabin numarası. Bu değişken genellikle çok fazla eksik veri içerir, çünkü sadece birinci sınıf yolcuların kabin bilgileri daha sık kaydedilmiştir.

12. **Embarked**: Yolcunun gemiye bindiği liman.
   - C: Cherbourg
   - Q: Queenstown
   - S: Southampton

```python
# Gerekli kütüphaneleri içe aktar
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew


url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Sayısal değişkenler
numerical_vars = ['Age', 'Fare']

# Eksik veri kontrolü
print("Eksik Veriler:")
print(df[numerical_vars].isnull().sum())

# Görselleştirme için grafik ayarları
plt.figure(figsize=(12, 5))

# Her değişken için histogram ve KDE grafiği çiz
for i, var in enumerate(numerical_vars, 1):
    plt.subplot(1, 2, i)
    sns.histplot(df[var].dropna(), kde=True, bins=30, color='skyblue')
    plt.title(f'{var}')
    plt.xlabel(var)
    plt.ylabel('Frekans')

plt.tight_layout()
plt.show()
```

Çarpıklık (Skewness): Ortalama > Medyan → Sağa çarpık dağılım. Tersi sola çarpık.

---

### Fare Değişkeninin İstatistiksel Özeti
Titanic veri setindeki `Fare` değişkeni, yolcuların ödediği bilet ücretlerini temsil eder. Örnek istatistikler şu şekilde:

- **Ortalama (Mean)**: 32.2
- **Medyan (50%)**: 14.4
- **Maksimum (Max)**: 512
- **Minimum (Min)**: 0
- **Standart Sapma (Std)**: 49.7

---

### 1. Ortalama (Mean): 32.2
- **Ne anlama gelir?**: Tüm yolcuların bilet ücretlerinin toplamı, yolcu sayısına (891) bölündüğünde, ortalama bilet ücreti 32.2 dolar çıkıyor. Yani, "tipik" bir yolcu, yaklaşık 32 dolarlık bir bilet almış gibi görünüyor.
- **Neden önemli?**: Ortalama, veri setindeki tüm değerlerin genel bir özetini verir. Ancak, aykırı değerler (örneğin, çok pahalı biletler) ortalamayı yukarı çekebilir, bu yüzden tek başına yanıltıcı olabilir.
---

### 2. Medyan (50%): 14.4
- **Ne anlama gelir?**: Bilet ücretlerini küçükten büyüğe sıralarsak, tam ortadaki değer 14.4 dolar. Yani, yolcuların yarısı 14.4 dolardan daha az, yarısı daha fazla ödemiş.
- **Neden önemli?**: Medyan, ortalamadan farklı olarak aykırı değerlerden etkilenmez. Bu yüzden, verinin "tipik" değerini daha iyi temsil eder. `Fare` için medyanın 14.4 olması, çoğu yolcunun düşük ücretli (muhtemelen üçüncü sınıf) bilet aldığını gösteriyor.

---

### 3. Maksimum (Max): 512
- **Ne anlama gelir?**: Veri setindeki en yüksek bilet ücreti 512 dolar. Bu, muhtemelen birinci sınıf lüks bir süit için ödenmiş bir ücret.
- **Neden önemli?**: Maksimum değer, veri setindeki en uç noktayı gösterir. Eğer max değeri, diğer değerlerden (örneğin, medyan veya 75% çeyreklik) çok uzaksa, bu aykırı bir değer (outlier) olabilir.

---

### 4. Minimum (Min): 0
- **Ne anlama gelir?**: Bazı yolcular bilet için hiç para ödememiş. Bu, ücretsiz seyahat eden yolcular (örneğin, mürettebat, çocuklar veya promosyonlu biletler) olduğunu gösteriyor.
- **Neden önemli?**: Minimum değer, veri setindeki en düşük noktayı gösterir. 0 değeri, veri setinde ücretsiz biletlerin nadir ama mevcut olduğunu ifade eder.

---

### 5. Standart Sapma (Std): 49.7
- **Ne anlama gelir?**: Standart sapma, bilet ücretlerinin ortalamadan (32.2) ne kadar saptığını gösterir. 49.7 gibi yüksek bir standart sapma, ücretlerin çok çeşitli olduğunu ve bazılarının ortalamadan çok uzak olduğunu ifade eder.
- **Neden önemli?**: Standart sapma, verinin ne kadar "dağılmış" olduğunu ölçer. Düşük bir standart sapma, değerlerin ortalamaya yakın olduğunu; yüksek bir standart sapma ise değerlerin geniş bir aralıkta dağıldığını gösterir.
---
---

### Genel Yorumlama
`Fare` değişkeninin istatistiksel özeti, bize şunu söylüyor:
- **Çoğu yolcu düşük ücretli biletler almış**: Medyan (14.4) çoğu yolcunun üçüncü sınıf veya ucuz ikinci sınıf bilet aldığını gösteriyor.
- **Birkaç yolcu çok pahalı biletler almış**: Maksimum (512) ve yüksek ortalama (32.2), az sayıda birinci sınıf yolcunun lüks biletler için yüksek ücretler ödediğini gösteriyor.
- **Dağılım sağa çarpık**: Ortalama > Medyan (32.2 > 14.4) ve yüksek çarpıklık birkaç yüksek ücretin sağ kuyruğu uzattığını teyit ediyor.
- **Veri çok oynak**: Yüksek standart sapma (49.7), bilet ücretlerinin çok çeşitli olduğunu ve farklı sınıflardaki yolcular arasında büyük fiyat farkları olduğunu gösteriyor.
- **Aykırı değerler var**: 512 dolar gibi uç değerler, aykırı olarak öne çıkıyor ve ortalamayı şişiriyor.

**groupby() ile Kategorik Özetleme Fonksiyonu**
kategorik değişken ile hedef değişkeni ilişkilendirip otomatik analiz sunar:


```python
def cat_summary(dataframe, col_name, target="Survived"):
    print(dataframe.groupby(col_name)[target].agg(["count", "mean", "median"]))
```

```python
cat_summary(df, "Embarked")
```

**groupby() Kullanım Alanları**

| Kullanım Amacı                          | Örnek                                          |
| --------------------------------------- | ---------------------------------------------- |
| Gruplara göre ortalama                  | `df.groupby("Sex")["Survived"].mean()`         |
| Çoklu istatistik almak                  | `.agg(["mean", "min", "max"])`                 |
| Çoklu sütuna göre gruplama              | `groupby(["Sex", "Pclass"])`                   |
| Segmentasyon ve pivot hazırlıkları      | Pivot tablolar için temel oluşturur            |
| Hedef değişken ilişkilerini analiz etme | Kategorik değişkenlere göre `Survived` analizi |

```python
df.dtypes

# Sayısal ve kategorik değişkenleri ayıran fonksiyon
def grab_col_names(dataframe, cat_th=10, car_th=20):
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]
    cat_cols += num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    return cat_cols, num_cols, cat_but_car

cat_cols, num_cols, cat_but_car = grab_col_names(df)
```
